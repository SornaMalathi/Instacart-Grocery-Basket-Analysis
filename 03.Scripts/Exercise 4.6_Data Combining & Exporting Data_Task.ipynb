#Import Dataframes

```python
import pandas as pd
import numpy as np
import os
```

# Import the pickle format df_orders_products_combined Dataframe 

```python
path = r'C:\Users\sorna\Desktop\Data Analysis\Achievement 4'
```

```python
df_orders_products_combined = pd.read_pickle(os.path.join(path, '02.Data', 'Prepared Data', 'orders_products_combined.pkl'))
```

```python
df_orders_products_combined.shape
```

# Import the cleaned Products Dataframe 

```python
df_prods_cleaned = pd.read_csv(os.path.join(path, '02.Data', 'Prepared Data', 'product_cleaned.csv'), index_col = False)
```

```python
df_prods_cleaned.shape
```

```python
df_orders_products_combined
```

# Drop the merge column from the Dataframe inorder to perform the merge function to these Dataframes using the indicator = True arguement

```python
df_orders_products_combined = df_orders_products_combined.drop(['_merge'], axis = 1)
```

```python
df_orders_products_combined
```

```python
df_prods_cleaned
```

# Merge the two Dataframes using the INNER join and using the product_id as the key column.

```python
df_ords_prods_merged = df_prods_cleaned.merge(df_orders_products_combined, on = 'product_id', indicator = True)
```

```python
df_ords_prods_merged.shape
```

```python
df_ords_prods_merged['_merge'].value_counts()
```

# Created a new Dataframe using the OUTER join arguement to cross check the merged Dataframe.

```python
df_ords_prods_merged_1 = df_prods_cleaned.merge(df_orders_products_combined, on = 'product_id', indicator = True, how = 'outer')
```

```python
df_ords_prods_merged_1.shape
```

```python
df_ords_prods_merged_1['_merge'].value_counts()
```

# Export the merged Dataframe in the pickle format

```python
df_ords_prods_merged.to_pickle(os.path.join(path, '02.Data','Prepared Data', 'ords_prods_merged.pkl'))
```
